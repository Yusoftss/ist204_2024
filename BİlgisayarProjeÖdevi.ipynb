```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('Kitap1.xlsx')
print(df1)
```

```python
df3=df1.iloc[0:82, 0:85]
display(df3)
```

```python
df2 = df3.sort_values(by='İller', ascending=False)
top_cities = df2.head(10).sort_values(by='Toplam', ascending=True)
# Grafik oluşturma
plt.figure(figsize=(9, 6))
ax = sns.barplot(x=top_cities['İller'], y=top_cities['Toplam'], palette="Reds", alpha=0.85)
plt.title("2023 Yılında Toplam En Çok Göç Eden 10 İl", fontsize=15)
plt.xlabel("İller", fontsize=13)
plt.ylabel("Toplam Göç Eden Nüfus", fontsize=13)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
# Çubuklara etiket ekleme
for bar, label in zip(ax.patches, top_cities['Toplam']):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height - height * 0.00, f'{label:,}', ha='center', va='bottom', fontsize=10, color='black', rotation=0)
plt.show()
```

```python
df_top_10 = df15[df15['İl'].isin(en_buyuk_10_il)]

plt.figure(figsize=(9, 6))
sns.violinplot(x='İl', y='Net goç hizi', data=df_top_10, palette='husl')



plt.xlabel('İl', fontsize=12)
plt.ylabel('Net Göç Hızı', fontsize=12)
plt.title('En Yüksek 10 İlin Net Göç Hızı Dağılımı', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
pivot_df = df15.pivot_table(index="Yil", columns="İl", values="Net goç")

top_10_cities = pivot_df.sum().sort_values(ascending=False).head(10).index

pivot_df_top_10 = pivot_df[top_10_cities]

plt.figure(figsize=(9, 6))
heatmap = sns.heatmap(pivot_df_top_10, annot=True, cmap="coolwarm", fmt=".0f", linewidths=.5)
heatmap.set_title("En Büyük 10 Şehrin Net Göç Isı Haritası")
plt.xlabel("Şehirler")
plt.ylabel("Yıl")
plt.show()
```
